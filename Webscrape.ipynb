```python
# Check if an element is present
def is_element_present(by, value): 
    try:
        WebDriverWait(driver, 10).until(EC.presence_of_element_located((getattr(By, by), value)))
        return True
    except:
        return False

# Click the "Next" button if present
def next_page():
    try:
        # Wait for the "Next" button to be clickable and retry if it's stale
        next_button = WebDriverWait(driver, 10).until(
            EC.element_to_be_clickable((By.XPATH, "//a[contains(@class, '_9QVEpD') and span[text()='Next']]"))
        ).click()
    except:
        if is_element_present("XPATH", "//a[contains(@class, '_9QVEpD') and span[text()='Next']]"):
            # Retry if the element becomes stale
            next_page()

# Initialize data storage
data = {
    'rating': [],
    'review_summary': [],
    'review_description': [],
    'product_id': [],
    'Price': []
}
```

```python
# Navigate to the initial page 
for link in range(len(url)):  # Adjust the range based on the number of URLs
    driver.get(url[link])
    driver.implicitly_wait(1)
    
    if is_element_present("XPATH", "//div[@class='col pPAw9M']/a"):
        driver.find_element(By.XPATH, "//div[@class='col pPAw9M']/a").click()
        driver.implicitly_wait(1)

    # Extract multiple elements for rating, summary, and description
    rating_elements = driver.find_elements(By.XPATH, "//div[@class='EKFha-']/div/div/div[1]/div")
    review_summaries = driver.find_elements(By.XPATH, "//p[@class='z9E0IG']")
    review_descriptions = driver.find_elements(By.XPATH, "//div[@class='ZmyHeo']")
    driver.implicitly_wait(1)


    # Check if there are any reviews to extract
    while is_element_present("XPATH", "//a[contains(@class, '_9QVEpD') and span[text()='Next']]"):

        # Handle "Read more" button if it exists
        while is_element_present("XPATH", "//span[contains(@class, 'b4x-fr') and span[text()='READ MORE']]"):
            read_more = driver.find_element(By.XPATH, "//span[contains(@class, 'b4x-fr') and span[text()='READ MORE']]")
            read_more.click()
            driver.implicitly_wait(1)

        for i in range(max([len(rating_elements),len(review_summaries),len(review_descriptions)])):
            rating_elements = driver.find_elements(By.XPATH, "//div[@class='EKFha-']/div/div/div[1]/div")
            review_summaries = driver.find_elements(By.XPATH, "//p[@class='z9E0IG']")
            review_descriptions = driver.find_elements(By.XPATH, "//div[@class='ZmyHeo']")
            driver.implicitly_wait(1)
            
            # Extract the rating, summary, and description if available
            try:
                data['rating'].append(rating_elements[i].text)
            except Exception:
                data['rating'].append(np.nan)

            try:
                data['review_summary'].append(review_summaries[i].text)
            except Exception:
                data['review_summary'].append("Null")

            try:
                review_text = review_descriptions[i].text.replace("\n", ".")
                data['review_description'].append(review_text)
            except Exception:
                data['review_description'].append("Null")

            # Add product ID and price (assuming they are available)
            data['product_id'].append(product[link])  # Replace with correct product logic
            data['Price'].append(price[link])  # Replace with correct price logic
            driver.implicitly_wait(1)

        # Move to the next page
        driver.implicitly_wait(1)
        
        # Stop if there are no more review pages
        if is_element_present("XPATH", "//a[contains(@class, '_9QVEpD') and span[text()='Next']]"):
            try:
                # Wait for the "Next" button to be clickable and retry if it's stale
                next_button = WebDriverWait(driver, 10).until(
                    EC.element_to_be_clickable((By.XPATH, "//a[contains(@class, '_9QVEpD') and span[text()='Next']]"))
                ).click()
                driver.implicitly_wait(1)
            except:
                next_page()
        else:
            break
```
